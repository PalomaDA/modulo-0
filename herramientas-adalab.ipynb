# **ESPACIO DE TRABAJO ADALAB**

### Flipped classroom
- Prepara la lección en casa (estará en git)
- Evaluación previa de conocimientos
- Ejercicios prácticos en pareja (se hacen en horario de clase)

### Proyectos
- Equipos de 4 o 5 alumnas
- *Agile & Scrum*
- Proyecto de empresa

### Aprendizaje centrado en la alumna
- Responsable de mi aprendizaje
- Rol del profesor: facilitador
- Se van a pedir mínimos pero no máximos (*"el cielo es el límite"*)
- Aprendizaje compartido entre todas (*"Mi código no es mío, es de la comunidad"*)
- Exponer mis errores es lo mejor que puedo hacer

### Programa formativo profeisonal
- Adalab = trabajo
- Actitud profesional
- Asistencia / conexión obligatoria y compromiso

### Importante:
- Autonomía: Ante la duda, buscar recursos (San Google) y preguntar a profesores y compañeras
- Colaboración: Ayuda a crear y mejorar nuevos materiales (Ayudanos a ayudarte)

### Trabajo en equipo y comunicación
- Canales apropiados
- Gestión de conflictos
- Hay que colaborar, no competir
- Aprendizaje compartido, entre todas manetenemos el nivel de clase igualado

### Gestión del tiempo
- Puntualidad
- Descansos
- Respeto por todas las facilitadoras

> Gestión de la frustración: La frustración es normal en el aprendizaje.

### Normas de convivencia en remoto
- Comunicación de ausencias justificadas/imprevistos y dudas a tu profe de referencia
- Puntualidad a la hora de conectarse (5 min de cortesía)
Escucha y respeto. SIlenciar los micrófonos cuando no estemos hablando
Webcam encedida
- Hacer descansos en los minutos marcados (son muchas horas delante del ordenador)

### Horario Adalab

### Tipos de clase:

    - Se priorizan las dudas básicas
    - empujamos el nivel de la clase desde abajo para que nadie se quede atrás
    - Preparación de las evaluaciones

## ALIANZA PROMO D
- Trabajo en equipo
- Confiar en las compañeras
- Compartir nuestro aprenedizaje
- Ser solidarias, comunicación entre compañeras. **Asertividad**
    - Vigilar cómo nos quejamos, cómo nos comunicamos con las compañeras - *Técnica del sandwitch*
- Empatía, escucha activa y paciencia con las compañeras. Respeto ante la diversidad de puntos fuertes y débiles
- Sentido del humor (memes)
- No gritarnos y no pisarnos unas a otras cuando estamos hablando, respetar el turno de palabra
- Decir las cosas buenas
- Darnos apoyo en los momentos de frustración o desánimo
- Llegar a acuerdos y cumplirlos. *Ir todas a una*
- **Ser positivas**. Confiar en nosotras mismas
